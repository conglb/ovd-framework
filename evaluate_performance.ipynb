```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m1 = './data-collection/performance_log.log'
m2 = './data-cleaning/performance_log.log'
m3 = './data-storage/performance_log.log'
m4 = './data-presentation/performance_log.log'

df1 = pd.read_csv(m1)
df2 = pd.read_csv(m2)
df3 = pd.read_csv(m3)
df4 = pd.read_csv(m4)
df = pd.DataFrame()
```

```python
df1 = df1[:21]
df2 = df2[:21]
df3 = df3[:8]
df4 = df4[:13]
df4
```

```python
df['m1_Cumulative_File_Size'] = df1['File_size'].cumsum() / (1024 ** 2)
df['m1_Cumulative_Time'] = df1['Time'].cumsum() 
# Scale the data
max_file_size = 1 * 1024**3  # 1GB in bytes
max_time = 10  # 10 seconds


# Plot the scaled cumulative data
plt.figure(figsize=(8, 6))
plt.plot(df['m1_Cumulative_File_Size'], df['m1_Cumulative_Time'], marker='o', linestyle='-', color='b')
plt.xlabel('Cumulative File Size (scaled to 1GB)', fontsize=12)
plt.ylabel('Cumulative Time Taken (scaled to 10s)', fontsize=12)
plt.title('Cumulative File Size vs Time Taken to Process (Scaled)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
df['m2_Cumulative_File_Size'] = df2['File_size'].cumsum() / (1024 ** 2)
df['m2_Cumulative_Time'] = df2['Time'].cumsum()
# Scale the data
max_file_size = 1024**2  # 1GB in bytes
max_time = 10  # 10 seconds


# Plot the scaled cumulative data
plt.figure(figsize=(8, 6))
plt.plot(df['m1_Cumulative_File_Size'], df['m2_Cumulative_Time'], marker='o', linestyle='-', color='b')
plt.xlabel('Cumulative File Size (scaled to 1GB)', fontsize=12)
plt.ylabel('Cumulative Time Taken (scaled to 10s)', fontsize=12)
plt.title('Cumulative File Size vs Time Taken to Process (Scaled)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
df['m3_Cumulative_File_Size'] = df3['File_size'].cumsum() / (1024 ** 2)
df['m3_Cumulative_Time'] = df3['Time'].cumsum() 
# Scale the data
max_file_size = 1024**2  # 1GB in bytes
max_time = 10  # 10 seconds


# Plot the scaled cumulative data
plt.figure(figsize=(8, 6))
plt.plot(df['m3_Cumulative_File_Size'], df['m3_Cumulative_Time'], marker='o', linestyle='-', color='b')
plt.xlabel('Cumulative File Size (scaled to 1GB)', fontsize=12)
plt.ylabel('Cumulative Time Taken (scaled to 10s)', fontsize=12)
plt.title('Cumulative File Size vs Time Taken to Process (Scaled)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
df['m4_Cumulative_File_Size'] = df4['File_size'].cumsum() / (1024 ** 2)
df['m4_Cumulative_Time'] = df4['Time'].cumsum() 

# Scale the data
max_file_size = 1 * 1024**3  # 1GB in bytes
max_time = 10  # 10 seconds


# Plot the scaled cumulative data
plt.figure(figsize=(8, 6))
plt.plot(df['m4_Cumulative_File_Size'], df['m4_Cumulative_Time'], marker='o', linestyle='-', color='b')
plt.xlabel('Cumulative File Size (scaled to 1GB)', fontsize=12)
plt.ylabel('Cumulative Time Taken (scaled to 10s)', fontsize=12)
plt.title('Cumulative File Size vs Time Taken to Process (Scaled)', fontsize=14)
plt.grid(True)
plt.show()
```

```python
fig = plt.figure(figsize=(8, 6))
plt.plot(df['m1_Cumulative_File_Size'], df['m1_Cumulative_Time'], marker='o', linestyle='-', color='g', label='Module Data Collection')
plt.plot(df['m2_Cumulative_File_Size'], df['m2_Cumulative_Time'], marker='o', linestyle='-', color='b', label='Module Data Cleaning')
plt.plot(df['m3_Cumulative_File_Size'], df['m3_Cumulative_Time'], marker='o', linestyle='-', color='purple', label='Module Data Storage')
plt.plot(df['m4_Cumulative_File_Size'], df['m4_Cumulative_Time'], marker='o', linestyle='-', color='r', label='Module Data Presentation')
plt.xlabel('Data Size (MB)', fontsize=12)
plt.ylabel('Time Taken (s)', fontsize=12)
plt.title('Cumulative Data Size vs Time Taken to Process', fontsize=14)
plt.legend()  # Add this to display legends
plt.grid(True)
plt.show()
fig.savefig("evaluation_modules.png")
```
